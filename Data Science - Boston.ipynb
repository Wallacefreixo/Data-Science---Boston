```python
boston = load_boston()
```

```python
#Mostre os primeiros registros da tabela
data = pd.DataFrame(boston.data, columns=[boston.feature_names])
data = data.assign(target=pd.Series(boston.target))
#primeiras 5 linhas
data.head()
```

```python
#Prepare X usando o pandas
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']

# use the list to select a subset of the original DataFrame
X = data[feature_cols]

# equivalent command to do this in one line
X = data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']]

# print the first 5 rows
X.head()
```

```python
#Qual o tipo de dados de X?
print(type(X))
print(X.shape)
```

```python
#Prepare Y usando o pandas
y = data['target']

# equivalent command that works if there are no spaces in the column name
y = data.target

# print the first 5 values
y.head()
```

```python
#Qual o tipo de dados de Y?
print(type(y))
print(y.shape)
```

```python
#Sobre o que se trata essa base de dados? Que tipo de informações ela guarda?
print(boston.DESCR)
```

```python
#Gere um X de treino e y de treino, X de teste e y da base (Split Train/Test)
from sklearn.cross_validation import train_test_split
```

```python
#importante test-size é a porcentagem de teste, nesse caso foi 40%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
```

```python
# se o teste deu 40% então o treino é 60%
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Usando modelo de regressão linear do sklearn, treine o modelo com o X e y de treino
# importar modelo
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train, y_train)
```

```python
#imprimir os atributos de "intercept" e "coefficients" que foram gerados
print(linreg.intercept_)
print(linreg.coef_)
```

```python
#imprimir o par "feature names" com os "coefficients"
list(zip(feature_cols, linreg.coef_))
```

```python
#Faça uma previsão usando o conjunto de X de teste
y_pred = linreg.predict(X_test)
```

```python
# define true and predicted response values
true = [100, 50, 30, 20]
pred = [90, 50, 50, 30]
```

```python
# calculate MAE by hand# calcul 
print((10 + 0 + 20 + 10)/4.)

# calculate MAE using scikit-learn
from sklearn import metrics
print(metrics.mean_absolute_error(true, pred))
```

```python
# calculate MSE by hand
print((10**2 + 0**2 + 20**2 + 10**2)/4.)

# calculate MSE using scikit-learn
print(metrics.mean_squared_error(true, pred))
```

```python
#Mostre a margem de error, usando o método "Root Mean Squared Error (RMSE)
import numpy as np
print(np.sqrt((10**2 + 0**2 + 20**2 + 10**2)/4.))

# calculate RMSE using scikit-learn
print(np.sqrt(metrics.mean_squared_error(true, pred)))
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#Existe uma forma de melhorar essa margem para que fique menor? Se sim, como seria?
# create a Python list of feature names
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']

# use the list to select a subset of the original DataFrame
X = data[feature_cols]

# select a Series from the DataFrame
y = data.target

# split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train, y_train)

# make predictions on the testing set
y_pred = linreg.predict(X_test)

# compute the RMSE of our predictions
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
